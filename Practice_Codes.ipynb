```python
import math
def vcpsubnet(ip, size):
    if size == 1:
        strip = ip.split("/")
        strip = strip[0]
        strip = strip.split(".")
        if strip[3] == '255':
            strip[2] = int(strip[2]) + 1
            strip[3] = int(0)
        else:
            strip[2] = int(strip[2]) + 1
        end = ".".join(str(x) for x in strip)  
        return end
    else:
        strip = ip.split("/")
        strip = strip[0]
        strip = strip.split(".")
        if size % 256 == 0:
            strip[3] = int(255)
        strip[2] = int(strip[2]) + (int(size/255)-1)
        start = ip + '/' + str(32-int(math.log(size,2)))
        end = ".".join(str(x) for x in strip) + '/' + str(32-int(math.log(size,2)))
        return start, end
```
